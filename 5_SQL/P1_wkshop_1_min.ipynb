```python
# 2.2.6 Display the minimum, maximum and average popularity scores. 
# The output of your query should contain only one row with 3 columns.
pd.read_sql_query("""SELECT MIN(song_popularity) as minimum, MAX(song_popularity) as maximum, AVG(song_popularity) as average FROM songs WHERE song_popularity = 
                        (SELECT MIN(song_popularity) FROM songs) 
                        OR 
                        (SELECT MAX(song_popularity) FROM songs)
                        OR 
                        (SELECT AVG(song_popularity) FROM songs  )
                        """,conn1)
```

```python
# 2.2.7 Display average popularity scores and count for songs that contain "love" in their name
# and those that don't. Is there any difference in their popularity?
pd.read_sql_query('''SELECT AVG(song_popularity) as popularity_with_love, COUNT(song_name) as numbers_of_songs FROM songs WHERE song_name LIKE '%love%' ''', conn1)
```

```python
pd.read_sql_query('''SELECT AVG(song_popularity) as popularity_no_love, COUNT(song_name) FROM songs WHERE song_name NOT LIKE '%love%' ''',conn1)
```

```python
# There are less songs with 'LOVE' in their title, but they are more popular on average
```
