```python
dfs = {}
def make_dfs_n_tables(dir_name = os.getcwd()):
    for file in os.listdir(dir_name):
        path = os.path.join(dir_name, file)
        if os.path.isfile(path) and (path[-4:]=='.csv'):
            # Get the DF
            df = pd.read_csv(path, header=0)
            dfs[file[:-4]]=df
            
            # Create Table from the DF into the baseballDB.sqlite
            try:
                
                df.to_sql(file[:-4],con=conn,index=False,if_exists='replace')
                print(f"{file[:-4]} : table creation successful")
            except Error as e:
                print("The error % occurred.".format(e))
            
```

```python
# 1.1 Which player has had the most homeruns? List top 25 players and provide their first and last names
# according to the file readme2014.txt, the homeruns are listed in the table 'BattingPost', under the attribute 'HR'
pd.read_sql_query('''
SELECT t1.HR, t2.playerID, t2.nameFirst as FirstName, t2.nameLast as LastName
FROM BattingPost t1
LEFT JOIN People t2
ON t1.playerID = t2.playerID
ORDER BY HR DESC LIMIT 25''',conn)
```

```python
# 1.2 Which team historically has paid the highest salary on average?
# Your query should only output teams whose average salary is above $1M.
# If average salary is formatted as scientific notation, convert it to numbers, i.e. 4.291454e+06 = 4291453

```

```python
pd.read_sql_query('''
SELECT CAST(AVG(t2.salary) as INT) as average_Salary, t2.teamID as team, t1.name
FROM teams t1
INNER JOIN Salaries t2
ON t1.teamID = t2.teamID
GROUP BY t2.teamID
HAVING AVG(t2.salary)>1000000
''',conn)
```

```python
# 1.3 Which team has the most players in the Hall of Fame?

# t2 = HallOfFame : playerID
# t1 = salaries : teamID | playerID 
# t3 = Teams : teamID 
# People : playerID

# We are going to use the salaries table, to connect playerID (from t2) to teamID (from t3)
# We link each player of the HallOfFame to its Team
# We group by team
# We count the distinct occurrences of a player appearing in the HOF
# we sort out by DESC
# we take the top team

pd.read_sql_query('''
SELECT COUNT(DISTINCT(t2.playerID)) as occurrence, t3.teamID as team_hallOfFame
FROM salaries t1
JOIN HallOfFame t2
ON t2.playerID = t1.playerID
JOIN Teams t3
ON t3.teamID = t1.teamID
GROUP BY t3.teamID
ORDER BY occurrence DESC
LIMIT 1
''',conn)
```

```python
# 1.4 Is there a relation between how many homeruns a player has made in a year and their salary that year? 
# Pull Player's name, ID, Year, Salary and homeruns together in a single query.

# GOOGLE : Definition of a homerun
# " A home run occurs when a batter hits a fair ball and scores on the play without being put out 
#  or without the benefit of an error. 
#  In almost every instance of a home run, a batter hits the ball in the air over the outfield fence 
# in fair territory."

# t1 = salaries : playerID 
# t2 = batting : playerID | HRs

df = pd.read_sql_query('''
SELECT t2.yearID, t2.playerID, t2.HR, SUM(t1.salary) as annual_salary
FROM salaries t1
INNER JOIN Batting t2
ON t1. playerID = t2.playerID
WHERE t2.HR !=0
GROUP BY t2.yearID, t2.playerID
ORDER BY t2.yearID
''',conn)
df
```

```python
# Now let's see what the correlation is between HR vs salary
print('The correlation between a Homerun and the player annual salary is:',round(df['HR'].corr(df['annual_salary']),2))
```
